# Ejercicio 5: Interpolación Lineal con Resultados del Ejercicio 3

Utilice los resultados del ejercicio 3 y la interpolación lineal para aproximar los siguientes valores de y(t). Compare las aproximaciones asignadas para los valores reales obtenidos mediante las funciones determinadas en el ejercicio 4.

a. y(0.25) y y(0.93)
b. y(1.25) y y(1.93)
c. y(2.10) y y(2.75)
d. y(0.54) y y(0.94)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def euler_method(f, t0, y0, h, t_end):
    """
    Implementación del método de Euler
    """
    t_values = [t0]
    y_values = [y0]
    
    t = t0
    y = y0
    
    while t < t_end:
        y_new = y + h * f(t, y)
        t_new = t + h
        
        t_values.append(t_new)
        y_values.append(y_new)
        
        t = t_new
        y = y_new
    
    return np.array(t_values), np.array(y_values)

def linear_interpolation(t_data, y_data, t_target):
    """
    Interpolación lineal para encontrar y en t_target
    """
    # Encontrar los puntos más cercanos
    idx = np.searchsorted(t_data, t_target)
    
    if idx == 0:
        return y_data[0]
    elif idx >= len(t_data):
        return y_data[-1]
    else:
        # Interpolación lineal entre idx-1 e idx
        t1, t2 = t_data[idx-1], t_data[idx]
        y1, y2 = y_data[idx-1], y_data[idx]
        
        # y = y1 + (y2-y1) * (t-t1)/(t2-t1)
        y_interp = y1 + (y2 - y1) * (t_target - t1) / (t2 - t1)
        return y_interp
```

## Obtener resultados del Ejercicio 3 (Método de Euler)

```python
# Problema a: y' = y/t - (y/t)²
def f_a(t, y):
    return (y/t) - (y/t)**2

t0_a, y0_a, h_a, t_end_a = 1, 1, 0.1, 4
t_euler_a, y_euler_a = euler_method(f_a, t0_a, y0_a, h_a, t_end_a)

# Problema b: y' = 1 + y/t + (y/t)²
def f_b(t, y):
    return 1 + (y/t) + (y/t)**2

t0_b, y0_b, h_b, t_end_b = 1, 0, 0.2, 3
t_euler_b, y_euler_b = euler_method(f_b, t0_b, y0_b, h_b, t_end_b)

# Problema c: y' = -(y+1)(y+3)
def f_c(t, y):
    return -(y + 1) * (y + 3)

t0_c, y0_c, h_c, t_end_c = 0, -2, 0.2, 2
t_euler_c, y_euler_c = euler_method(f_c, t0_c, y0_c, h_c, t_end_c)

# Problema d: y' = -5y + 5t² + 2t
def f_d(t, y):
    return -5*y + 5*t**2 + 2*t

t0_d, y0_d, h_d, t_end_d = 0, 1/3, 0.1, 1
t_euler_d, y_euler_d = euler_method(f_d, t0_d, y0_d, h_d, t_end_d)

print("Datos del Ejercicio 3 obtenidos exitosamente")
print(f"Problema a: {len(t_euler_a)} puntos de t=1 a t=4")
print(f"Problema b: {len(t_euler_b)} puntos de t=1 a t=3")
print(f"Problema c: {len(t_euler_c)} puntos de t=0 a t=2")
print(f"Problema d: {len(t_euler_d)} puntos de t=0 a t=1")
```

## Definir las soluciones exactas del Ejercicio 4

```python
# Soluciones exactas del Ejercicio 4
def exact_a(t):
    return t / (np.log(t) + 1)

def exact_b(t):
    return t * np.tan(np.log(t))

def exact_c(t):
    return -3 + 2 / (1 + np.exp(2*t))

def exact_d(t):
    return t**2 + (1/3) * np.exp(-5*t)

print("Soluciones exactas definidas")
```

## Problema a: y(0.25) y y(0.93)

```python
print("PROBLEMA A: y' = y/t - (y/t)²")
print("Solución exacta: y(t) = t/(ln(t) + 1)")
print("=" * 60)

# Nota: Los valores 0.25 y 0.93 están fuera del dominio [1,4] del problema a
# Necesitamos extender la solución hacia atrás para t < 1

# Extender hacia atrás desde t=1
t_ext_a = np.arange(0.1, 1, 0.1)
y_ext_a = []

# Usar Euler hacia atrás (h negativo)
t_back = 1.0
y_back = 1.0
h_back = -0.1

t_extended = [t_back]
y_extended = [y_back]

while t_back > 0.1:
    y_new = y_back + h_back * f_a(t_back, y_back)
    t_back += h_back
    t_extended.insert(0, t_back)
    y_extended.insert(0, y_new)
    y_back = y_new

# Combinar datos extendidos con originales
t_complete_a = np.concatenate([np.array(t_extended), t_euler_a[1:]])
y_complete_a = np.concatenate([np.array(y_extended), y_euler_a[1:]])

# Interpolación para los valores solicitados
targets_a = [0.25, 0.93]
results_a = []

for t_target in targets_a:
    y_interp = linear_interpolation(t_complete_a, y_complete_a, t_target)
    y_exact = exact_a(t_target)
    error = abs(y_exact - y_interp)
    error_rel = (error / abs(y_exact)) * 100
    
    results_a.append({
        't': t_target,
        'Interpolación': y_interp,
        'Exacta': y_exact,
        'Error Abs': error,
        'Error %': error_rel
    })
    
    print(f"t = {t_target}:")
    print(f"  Interpolación: {y_interp:.6f}")
    print(f"  Exacta:        {y_exact:.6f}")
    print(f"  Error abs:     {error:.6f}")
    print(f"  Error rel:     {error_rel:.3f}%")
    print()

df_results_a = pd.DataFrame(results_a)
print("Resumen:")
print(df_results_a.to_string(index=False, float_format='%.6f'))
```

## Problema b: y(1.25) y y(1.93)

```python
print("\nPROBLEMA B: y' = 1 + y/t + (y/t)²")
print("Solución exacta: y(t) = t·tan(ln(t))")
print("=" * 60)

# Valores a interpolar
targets_b = [1.25, 1.93]
results_b = []

for t_target in targets_b:
    y_interp = linear_interpolation(t_euler_b, y_euler_b, t_target)
    y_exact = exact_b(t_target)
    error = abs(y_exact - y_interp)
    error_rel = (error / abs(y_exact)) * 100
    
    results_b.append({
        't': t_target,
        'Interpolación': y_interp,
        'Exacta': y_exact,
        'Error Abs': error,
        'Error %': error_rel
    })
    
    print(f"t = {t_target}:")
    print(f"  Interpolación: {y_interp:.6f}")
    print(f"  Exacta:        {y_exact:.6f}")
    print(f"  Error abs:     {error:.6f}")
    print(f"  Error rel:     {error_rel:.3f}%")
    print()

df_results_b = pd.DataFrame(results_b)
print("Resumen:")
print(df_results_b.to_string(index=False, float_format='%.6f'))
```

## Problema c: y(2.10) y y(2.75)

```python
print("\nPROBLEMA C: y' = -(y+1)(y+3)")
print("Solución exacta: y(t) = -3 + 2/(1 + e^(2t))")
print("=" * 60)

# Para t = 2.75, necesitamos extender más allá de t = 2
# Extender la solución hasta t = 3
t_ext_c = t_euler_c[-1]
y_ext_c = y_euler_c[-1]
h_ext_c = 0.2

t_extended_c = list(t_euler_c)
y_extended_c = list(y_euler_c)

while t_ext_c < 3.0:
    y_new = y_ext_c + h_ext_c * f_c(t_ext_c, y_ext_c)
    t_ext_c += h_ext_c
    t_extended_c.append(t_ext_c)
    y_extended_c.append(y_new)
    y_ext_c = y_new

t_complete_c = np.array(t_extended_c)
y_complete_c = np.array(y_extended_c)

# Valores a interpolar
targets_c = [2.10, 2.75]
results_c = []

for t_target in targets_c:
    y_interp = linear_interpolation(t_complete_c, y_complete_c, t_target)
    y_exact = exact_c(t_target)
    error = abs(y_exact - y_interp)
    error_rel = (error / abs(y_exact)) * 100 if abs(y_exact) > 1e-10 else 0
    
    results_c.append({
        't': t_target,
        'Interpolación': y_interp,
        'Exacta': y_exact,
        'Error Abs': error,
        'Error %': error_rel
    })
    
    print(f"t = {t_target}:")
    print(f"  Interpolación: {y_interp:.6f}")
    print(f"  Exacta:        {y_exact:.6f}")
    print(f"  Error abs:     {error:.6f}")
    print(f"  Error rel:     {error_rel:.3f}%")
    print()

df_results_c = pd.DataFrame(results_c)
print("Resumen:")
print(df_results_c.to_string(index=False, float_format='%.6f'))
```

## Problema d: y(0.54) y y(0.94)

```python
print("\nPROBLEMA D: y' = -5y + 5t² + 2t")
print("Solución exacta: y(t) = t² + (1/3)e^(-5t)")
print("=" * 60)

# Valores a interpolar
targets_d = [0.54, 0.94]
results_d = []

for t_target in targets_d:
    y_interp = linear_interpolation(t_euler_d, y_euler_d, t_target)
    y_exact = exact_d(t_target)
    error = abs(y_exact - y_interp)
    error_rel = (error / abs(y_exact)) * 100
    
    results_d.append({
        't': t_target,
        'Interpolación': y_interp,
        'Exacta': y_exact,
        'Error Abs': error,
        'Error %': error_rel
    })
    
    print(f"t = {t_target}:")
    print(f"  Interpolación: {y_interp:.6f}")
    print(f"  Exacta:        {y_exact:.6f}")
    print(f"  Error abs:     {error:.6f}")
    print(f"  Error rel:     {error_rel:.3f}%")
    print()

df_results_d = pd.DataFrame(results_d)
print("Resumen:")
print(df_results_d.to_string(index=False, float_format='%.6f'))
```

```python
# Crear gráficos mostrando los puntos de interpolación
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Problema a
t_fine_a = np.linspace(0.1, 4, 1000)
y_exact_fine_a = exact_a(t_fine_a)

axes[0,0].plot(t_fine_a, y_exact_fine_a, 'r-', label='Exacta', linewidth=2)
axes[0,0].plot(t_complete_a, y_complete_a, 'bo-', label='Euler', markersize=3, alpha=0.7)
for result in results_a:
    axes[0,0].plot(result['t'], result['Interpolación'], 'gs', markersize=10, label=f"Interp t={result['t']}")
    axes[0,0].plot(result['t'], result['Exacta'], 'r*', markersize=12)
axes[0,0].set_title('a) Interpolación para y = t/(ln(t)+1)')
axes[0,0].legend()
axes[0,0].grid(True)
axes[0,0].set_xlabel('t')
axes[0,0].set_ylabel('y')

# Problema b
t_fine_b = np.linspace(1, 3, 1000)
y_exact_fine_b = exact_b(t_fine_b)

axes[0,1].plot(t_fine_b, y_exact_fine_b, 'r-', label='Exacta', linewidth=2)
axes[0,1].plot(t_euler_b, y_euler_b, 'bo-', label='Euler', markersize=6, alpha=0.7)
for result in results_b:
    axes[0,1].plot(result['t'], result['Interpolación'], 'gs', markersize=10, label=f"Interp t={result['t']}")
    axes[0,1].plot(result['t'], result['Exacta'], 'r*', markersize=12)
axes[0,1].set_title('b) Interpolación para y = t·tan(ln(t))')
axes[0,1].legend()
axes[0,1].grid(True)
axes[0,1].set_xlabel('t')
axes[0,1].set_ylabel('y')

# Problema c
t_fine_c = np.linspace(0, 3, 1000)
y_exact_fine_c = exact_c(t_fine_c)

axes[1,0].plot(t_fine_c, y_exact_fine_c, 'r-', label='Exacta', linewidth=2)
axes[1,0].plot(t_complete_c, y_complete_c, 'bo-', label='Euler', markersize=6, alpha=0.7)
for result in results_c:
    axes[1,0].plot(result['t'], result['Interpolación'], 'gs', markersize=10, label=f"Interp t={result['t']}")
    axes[1,0].plot(result['t'], result['Exacta'], 'r*', markersize=12)
axes[1,0].set_title('c) Interpolación para y = -3 + 2/(1+e^(2t))')
axes[1,0].legend()
axes[1,0].grid(True)
axes[1,0].set_xlabel('t')
axes[1,0].set_ylabel('y')

# Problema d
t_fine_d = np.linspace(0, 1, 1000)
y_exact_fine_d = exact_d(t_fine_d)

axes[1,1].plot(t_fine_d, y_exact_fine_d, 'r-', label='Exacta', linewidth=2)
axes[1,1].plot(t_euler_d, y_euler_d, 'bo-', label='Euler', markersize=4, alpha=0.7)
for result in results_d:
    axes[1,1].plot(result['t'], result['Interpolación'], 'gs', markersize=10, label=f"Interp t={result['t']}")
    axes[1,1].plot(result['t'], result['Exacta'], 'r*', markersize=12)
axes[1,1].set_title('d) Interpolación para y = t² + (1/3)e^(-5t)')
axes[1,1].legend()
axes[1,1].grid(True)
axes[1,1].set_xlabel('t')
axes[1,1].set_ylabel('y')

plt.tight_layout()
plt.show()
```
