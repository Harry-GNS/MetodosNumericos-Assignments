### Ejercicio 3

La siguiente tabla muestra los promedios de puntos del colegio de 20 especialistas en matemáticas y ciencias computacionales, junto con las calificaciones que recibieron estos estudiantes en la parte de matemáticas de la prueba ACT (Programa de Pruebas de Colegios Americanos) mientras estaban en secundaria. Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.

| Puntuación ACT | Promedio de puntos | Puntuación ACT | Promedio de puntos |
|---|---|---|---|
| 28    | 3.84    | 29    | 3.75    |
| 25    | 3.21    | 28    | 3.65    |
| 28    | 3.23    | 27    | 3.87    |
| 27    | 3.63    | 29    | 3.75    |
| 28    | 3.75    | 21    | 1.66    |
| 33    | 3.20    | 28    | 3.12    |
| 28    | 3.41    | 28    | 2.96    |
| 29    | 3.38    | 26    | 2.92    |
| 23    | 3.53    | 30    | 3.10    |
| 27    | 2.03    | 24    | 2.81    |

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del ejercicio 3
# Puntuaciones ACT (variable independiente x)
act = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 
                29, 28, 27, 29, 21, 28, 28, 26, 30, 24])

# Promedios de puntos (variable dependiente y)
gpa = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03,
                3.75, 3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81])

# Ajuste por mínimos cuadrados (regresión lineal)
coef = np.polyfit(act, gpa, 1) 
pendiente = coef[0]
intercepto = coef[1]

# Crear función de la recta ajustada
recta_ajustada = np.poly1d(coef)

# Calcular valores predichos
gpa_predicho = recta_ajustada(act)

# Calcular error cuadrático medio
error = np.mean((gpa - gpa_predicho)**2)

# Crear grafico
plt.figure(figsize=(10, 6))
plt.scatter(act, gpa, color='blue', label='Datos reales', alpha=0.7)

# Graficar la recta de regresión
x_vals = np.linspace(min(act)-1, max(act)+1, 100)
plt.plot(x_vals, recta_ajustada(x_vals), 'r-', 
         label=f'Regresión lineal: y = {pendiente:.3f}x + {intercepto:.3f}')

# Configuración del grafico
plt.title('Relación entre Puntuación ACT y Promedio de Puntos', fontsize=14)
plt.xlabel('Puntuación ACT en Matemáticas', fontsize=12)
plt.ylabel('Promedio de Puntos (GPA)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar ecuación y error
plt.text(22, 1.8, f'Ecuación: y = {pendiente:.4f}x + {intercepto:.4f}', 
         fontsize=10, bbox=dict(facecolor='white', alpha=0.8))
plt.text(22, 1.6, f'Error cuadrático medio: {error:.4f}', 
         fontsize=10, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()

# Resultados numéricos
print("\n== RESULTADOS DEL AJUSTE LINEAL ===")
print(f"Ecuación de la recta: y = {pendiente:.4f}x + {intercepto:.4f}")
print(f"Pendiente (coeficiente): {pendiente:.4f}")
print(f"Intercepto: {intercepto:.4f}")
print(f"Error cuadrático medio: {error:.4f}")
```
